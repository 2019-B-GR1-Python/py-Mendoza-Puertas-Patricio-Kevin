```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy import ndimage
from scipy import misc
from random import sample

mapache = misc.face()
rompecabeza = []
jugar = []

def cuatro_partes_rompecabeza(imagen):
    global rompecabeza
    arreglo1, arreglo2 = np.split(mapache,2)
    arreglo1, arreglo3 = np.hsplit(arreglo1,2)
    arreglo2, arreglo4 = np.hsplit(arreglo2,2)
    rompecabeza.append(arreglo1)
    rompecabeza.append(arreglo2)
    rompecabeza.append(arreglo3)
    rompecabeza.append(arreglo4)
    dibujar(rompecabeza)
    
    
def dibujar(posicion_pieza_rompecabeza):
    fig, axs = plt.subplots(2, 2)
    
    (ax2, ax1), (ax3, ax4) = axs
    
    axs[0, 0].imshow(posicion_pieza_rompecabeza[0])
    axs[0, 1].imshow(posicion_pieza_rompecabeza[1])
    axs[1, 0].imshow(posicion_pieza_rompecabeza[2])
    axs[1, 1].imshow(posicion_pieza_rompecabeza[3])
    plt.show
    rompecabeza_desordenado()
    
def rompecabeza_desordenado():
    global rompecabeza
    global jugar
    piezas_random = (sample(rompecabeza,4))
    longitud = len(rompecabeza)
    for i in range(longitud):
        jugar.append(piezas_random[i])

        
    
def armar_rompecabeza():
    global jugar
    aux_jugar = jugar[:]
    cambiar_pieza_1 = (int(input("Mover pieza: ")))
    cambiar_pieza_2 = (int(input("Mover la pieza seleccionada por la pieza: ")))
    jugar[cambiar_pieza_2] = jugar[cambiar_pieza_1]
    jugar[cambiar_pieza_1] = aux_jugar[cambiar_pieza_2]
    dibujar(jugar)
    

cuatro_partes_rompecabeza(mapache)
dibujar(jugar);
```

```python
armar_rompecabeza();
```
